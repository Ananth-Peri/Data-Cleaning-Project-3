```python
# Reading our csv files using a separator to separate the data given in single column into multiple respective columns

transactions = pd.read_csv("PD 2023 Wk 1 Input.csv", sep=',', engine='python')
targets = pd.read_csv("Targets.csv")
```

```python
# Splitting up the transaction code to obtain the bank code

transaction_new=transactions["Transaction Code"].str.split("-", n=1, expand=True); transaction_new
```

```python
# Adding the bank code column from the new dataframe to our existing "transactions" dataframe by creating a new column

transactions["Bank Code"] = transaction_new[0]; transactions
```

```python
# Subsetting our original dataframe to have data of "DSB"

transactions_DSB = transactions[(transactions["Bank Code"]=="DSB")]; transactions_DSB
```

```python
# Converting the data type of transaction date to datetime object and then converting to datetime string in the format below

transactions_DSB["Transaction Date"] = pd.to_datetime(transactions_DSB["Transaction Date"]).dt.strftime("%d-%m-%Y")
```

```python
# Updated dataframe

transactions_DSB.info()
```

```python
# Updated dataframe

transactions_DSB
```

```python
# We are replacing the values 1, 2 with respective values mentioned below

transactions_DSB["Online or In-Person"]=transactions_DSB["Online or In-Person"].replace(to_replace=[1,2],value=["Online","In-Person"])
```

```python
# Resetting the index of the dataframe

transactions_DSB.reset_index(inplace=True, drop=True)
```

```python
# Converting the transaction date to datetime object

transactions_DSB["Transaction Date"]=pd.to_datetime(transactions_DSB["Transaction Date"])
```

```python
# Creating a new column to convert the transaction date into respective year and quarter

# Q1 = Jan, Feb, Mar  Q2 = Apr, May, Jun  Q3 = Jul, Aug, Sept  Q4 = Oct, Nov, Dec

transactions_DSB["Quarter"] = transactions_DSB["Transaction Date"].dt.to_period("Q"); transactions_DSB
```

```python
# Dropping unnecessary columns

transactions_DSB.drop(columns=["Transaction Code","Customer Code", "Transaction Date","Bank Code"], inplace=True)
```

```python
# Creating a new dataframe by grouping all the values based on quarter and online or in-person and then summing up

result = transactions_DSB.groupby(["Online or In-Person","Quarter"])["Value"].agg("sum").reset_index(); result
```

```python
# Target dataframe

targets
```

```python
# Creating a new column in result by assigning column of targets

result["Expected Target"]= targets["Expected Value"]; result
```

```python
# Creating a new column to identify the difference between the expected target and the achieved value

result["Variance to Target"] = result["Value"]-result["Expected Target"]; result
```
